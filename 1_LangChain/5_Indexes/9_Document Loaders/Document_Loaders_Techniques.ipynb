📘 What are Document Loaders in RAG?
In LangChain, Document Loaders are utilities that help you load text data from various sources (PDFs, CSVs, URLs, Markdown, Notion, etc.) into a standard format (Document objects) for downstream processing like chunking, embedding, and retrieval.

🔧 Why are Document Loaders Important?
Unified format – All documents, no matter the source, are turned into Document objects.

Metadata retention – You can retain source info like file name, URL, author, page number, etc.

Flexible ingestion – Load from local files, APIs, databases, etc.

✅ Commonly Used Document Loaders in LangChain
Loader	Description
TextLoader -----> Loads plain text files (.txt)
PyPDFLoader ----> Loads content from PDFs
UnstructuredFileLoader ----> Smart parsing of HTML, Markdown, etc.
CSVLoader ----> Loads CSV files
WebBaseLoader ----> Loads text from URLs
NotionLoader ----> Loads pages from Notion
S3DirectoryLoader ----> Loads files from AWS S3
DirectoryLoader ----> Loads files from a folder

💡 Example 1: Load .txt File Using TextLoader

```python
from langchain.document_loaders import TextLoader

# Load a simple text file
loader = TextLoader("example.txt")
documents = loader.load()

# Print loaded documents
for doc in documents:
    print("Content:", doc.page_content)
    print("Metadata:", doc.metadata)
```

💡 Example 2: Load PDF Using PyPDFLoader

```python
from langchain.document_loaders import PyPDFLoader

# Load a PDF document
loader = PyPDFLoader("AttentionAllYouNeed.pdf")
documents = loader.load()

# Each page becomes a separate Document
for i, doc in enumerate(documents):
    print(f"Page {i+1} Content:\n{doc.page_content[:100]}...")
    print("Metadata:", doc.metadata)
    print("-" * 50)
```

💡 Example 3: Load Web Page Using WebBaseLoader

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://en.wikipedia.org/wiki/LangChain")
documents = loader.load()

for doc in documents:
    print("Page Content (excerpt):", doc.page_content[:200])
    print("Metadata:", doc.metadata)
```

💡 Example 4: Load Directory of Files

```python
from langchain.document_loaders import DirectoryLoader
from langchain.document_loaders import TextLoader

# Load all .txt files from a directory
loader = DirectoryLoader(
    path="docs/",
    glob="**/*.txt",
    loader_cls=TextLoader
)
documents = loader.load()

print(f"Loaded {len(documents)} documents.")
```
