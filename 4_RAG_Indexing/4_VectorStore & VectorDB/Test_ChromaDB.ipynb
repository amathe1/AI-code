```python
import chromadb
from chromadb.utils import embedding_functions

# Create client
client = chromadb.Client()

# Create collection
collection = client.create_collection(name="ai_docs")
```

```python
collection.add(
    ids=["doc_1"],
    documents=["AI is powerful system"],
    metadatas=[{"topic": "AI"}],
    embeddings=[[0.21, -0.33, 0.67, 0.05]]  # example vector
)
```

```python
results = collection.get(
    include=["embeddings", "documents", "metadatas"]
)

print(results)
```
